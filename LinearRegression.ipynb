```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the dataset
df=pd.read_csv('data.csv', encoding='ISO-8859-1')
```

```python
# Drop unnecessary columns
df.drop(columns=['InvoiceNo', 'CustomerID', 'StockCode', 'Description'], inplace=True)
```

```python
df=df[df['UnitPrice']>0 ]
df['TotalSales'] = df['Quantity'] * df['UnitPrice']
# Drop 'Quantity' and 'UnitPrice' columns as they are now represented by 'TotalSales'
df.drop(columns=['Quantity', 'UnitPrice'], inplace=True)
```

```python
# Convert 'StockCode', 'Description', and 'Country' to categorical data using one-hot encoding
df = pd.get_dummies(df, columns=['Country'], drop_first=True)
```

```python
# Feature engineering on InvoiceDate: Extracting useful time-based features
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['InvoiceYear'] = df['InvoiceDate'].dt.year
df['InvoiceMonth'] = df['InvoiceDate'].dt.month
df['InvoiceDay'] = df['InvoiceDate'].dt.day
df['InvoiceHour'] = df['InvoiceDate'].dt.hour
df.drop(columns=['InvoiceDate'], inplace=True)  # Drop the original column
```

```python
# Define features (X) and target (Y)
X = df.drop(columns=['TotalSales'])  # Using all columns except target
Y = df['TotalSales']  # Predicting UnitPrice
```

```python
# Split the dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Train the Linear Regression model
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Predict on the test set
Y_pred = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
```

```python
# Print evaluation metrics
print(f"Mean Squared Error: {mse}")
print(f"R-squared Value: {r2}")
```

```python
# display model coefficients
coefficients = pd.DataFrame({'Feature': X.columns, 'Coefficient': model.coef_})
print(coefficients)
```
